```python
import seaborn as sns
import matplotlib.pyplot as plt

def skewplot(resale_df, column):
    plt.figure(figsize=(20,5))

    plt.subplot(1,3,1)
    sns.boxenplot(data=resale_df, x=column,color='magenta')
    plt.title(f'Box Plot for {column}')

    plt.subplot(1,3,2)
    sns.histplot(data=resale_df, x=column, kde=True, bins=50, color="Maroon")
    plt.title(f'Distribution Plot for {column}')

    plt.subplot(1,3,3)
    sns.violinplot(data=resale_df, x=column, color='Chocolate')
    plt.title(f'Violin Plot for {column}')
    plt.show()

columns_to_plot = ['month', 'year', 'floor_area_sqm', 'storey_start', 'storey_end', 'resale_price','price_per_sqm']
for col in columns_to_plot:
    skewplot(resale_df, col)
```

OUTLIER HANDLING

```python
#creating histograms to visualize their distribution
col_out=["storey_start_log","storey_end_log","resale_price_log","price_per_sqm_log"]
```

```python
#calculate lower and upper bounds of an Interquartile Range (IQR) for outlier detection
def iqr_lower_bound(resale_df, col):
    Q1 = resale_df[col].quantile(0.25)
    Q3= resale_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5*IQR)
    return lower_bound
    
def iqr_upper_bound(resale_df, col):
    Q1 = resale_df[col].quantile(0.25)
    Q3= resale_df[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + (1.5*IQR)
    return upper_bound
```

```python
{f"upper bound of {i}": iqr_upper_bound(resale_df,i) for i in col_out}
```

```python
# Calculate the lower bounds for each column in col_out
{f"lower bound of {i}": iqr_lower_bound(resale_df, i) for i in col_out}
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

class RegressionModel:
    def __init__(self, df_last_final):
        self.df_last_final = df_last_final
        self.preprocess_data()
        self.split_data()
        self.initialize_models()
        
    def preprocess_data(self):
        self.x_reg = self.df_last_final.drop(columns="resale_price_log", axis=1)
        self.y_reg = self.df_last_final["resale_price_log"]
    
    def split_data(self):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.x_reg, self.y_reg,test_size=0.2,random_state=42)
        print(f'Train shape: {self.X_train.shape}')
        print(f'Test shape: {self.X_test.shape}')
        print(f'Train labels shape: {self.y_train.shape}')
        print(f'Test labels shape: {self.y_test.shape}')
    
    def initialize_models(self):
        self.models = {
            'LinearRegression': LinearRegression,
            'DecisionTreeRegressor': DecisionTreeRegressor,
            'RandomForestRegressor': RandomForestRegressor,
        }
    
    def evaluate_model(self, model_name, model_class):
        print(f"Training {model_name} model...")
        model = model_class().fit(self.X_train, self.y_train)
        print(f"Model {model_name} trained.")
        
        y_pred_train = model.predict(self.X_train)
        y_pred_test = model.predict(self.X_test)
        
        # Accuracy score for training data
        mse_train = mean_squared_error(self.y_train, y_pred_train)
        rmse_train = np.sqrt(mse_train)
        mae_train = mean_absolute_error(self.y_train, y_pred_train)
        r2_train = r2_score(self.y_train, y_pred_train)
        
        # Accuracy score for testing data
        mse_test = mean_squared_error(self.y_test, y_pred_test)
        rmse_test = np.sqrt(mse_test)
        mae_test = mean_absolute_error(self.y_test, y_pred_test)
        r2_test = r2_score(self.y_test, y_pred_test)
        
        metrics_train = {
            "Data": "Training Data",
            "Model": model_name,
            "R2_Score": r2_train,
            "Mean Absolute Error": mae_train,
            "Mean Squared Error": mse_train,
            "Root Mean Squared Error": rmse_train
        }
        metrics_test = {
            "Data": "Testing Data",
            "Model": model_name,
            "R2_Score": r2_test,
            "Mean Absolute Error": mae_test,
            "Mean Squared Error": mse_test,
            "Root Mean Squared Error": rmse_test
        }
        return metrics_train, metrics_test
    
    def evaluate_all_models(self):
        results = [self.evaluate_model(model_name, model_class) for model_name, model_class in self.models.items()]
        for metrics_train, metrics_test in results:
            print(metrics_train)
            print(metrics_test)
            print("")

model = RegressionModel(df_last_final)
model.evaluate_all_models()

```

```python
from sklearn.tree import DecisionTreeRegressor

x=df_last_final.drop(columns=['resale_price_log'],axis=1)
y=df_last_final['resale_price_log']

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size= 0.2,random_state=42)

decision_tree_model = DecisionTreeRegressor(max_depth=20)
decision_tree_model.fit(x_train, y_train)

y_pred_test = decision_tree_model.predict(x_test)

mse_test = mean_squared_error(y_test, y_pred_test)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)

print('Decision Tree Metrics')
print(f"RÂ² score: {r2_test}")
print(f"Mean Absolute Error: {mae_test}")
print(f"Mean Squared Error : {mse_test}")
print(f"Root Mean Squared Error: {rmse_test}")
```

```python
from sklearn.ensemble import RandomForestRegressor

x=df_last_final.drop(columns=['resale_price_log'],axis=1)
y=df_last_final['resale_price_log']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)

rf_model = RandomForestRegressor(n_estimators=150,max_depth=20)
rf_model.fit(x_train, y_train)

y_pred_test = rf_model.predict(x_test)

mse_test = mean_squared_error(y_test, y_pred_test)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)

print('Random Forest Metrics')
print(f"RÂ² score: {r2_test}")
print(f"Mean Absolute Error: {mae_test}")
print(f"Mean Squared Error : {mse_test}")
print(f"Root Mean Squared Error: {rmse_test}")


#Random Forest Metrics
#RÂ² score: 0.9999744790262044
#Mean Absolute Error: 0.00030113631383939693
#Mean Squared Error : 9.57825960720881e-06
#Root Mean Squared Error: 0.003094876347644411
```
